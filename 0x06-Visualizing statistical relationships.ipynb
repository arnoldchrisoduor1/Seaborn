# Visualizing Statistical Relationships
- Statistical analysis is the process of understanding how variables relate to each other in a dataset and how those relationships depend on other variables.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

## Relating variables with scatterplots.

```python
tips = sns.load_dataset("tips")
sns.relplot(data=tips, x="total_bill", y="tip")
```

```python
# we can add another dimension by coloring the points according to a third variable, this is called using a hue semantic.

sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker")
```

```python
# To emphasize the difference between the classes, and to improve accessibility, you can use a different marker style for each class:

sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", style="smoker"
)
```

```python
# It is also possible to represent four variables by changing the hue and style of each point independently
# This should be done carefu;y as the eye is much more sensitive to shape than color.

sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", style="time"
)
```

```python
# Default coloring switches to sequential palette if the hue was categorical:

sns.relplot(
    data=tips, x="total_bill", y="tip", hue="size"
)
```

```python
# we can also customize the string palette:

sns.relplot(
    data=tips,
    x="total_bill", y="tip",
    hue="size", palette="ch:r=-.5,l=.75"
)
```

```python
# The third kind of semantic variable changes the size of each point:

sns.relplot(data=tips, x="total_bill", y="tip", size="size")
```

```python
# we can customize the range of values in the data units into a range in area units:
sns.relplot(
    data = tips, x="total_bill", y="tip", size="size", sizes=(15, 200)
)
```

## Emphasizing continuity with line Plots

```python
# Using line plots to understand the variable of time:

dowjones = sns.load_dataset("dowjones")
sns.relplot(data=dowjones, x="Date", y="Price", kind="line")
```

### Aggragation and representing uncertainity
- More complex datasets might have different values for x, the default behaviour for x in seaborn is to aggregate the multiple measures at each x value by plotting the mean and the 95% confidence interval around the mean.

```python
fmri = sns.load_dataset("fmri")
sns.relplot(data=fmri, x="timepoint", y="signal", kind="line")
```

```python
# we can disable the computation of time intervals as they are done using the time intensive bootstrapping:

sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", errorbar=None,
)
```

```python
# Another good option is to represent the spread of distribution at each timepoint by plotting the standard deviation
# Instead of the confidence interval:

sns.relplot(
    data=fmri, kind="line", x="timepoint", y="signal", errorbar="sd"
)
```

```python
# To turn off aggregator we can turn off the estimator altogether giving us a starnge effect of data with multiple observations:

sns.relplot(
    data=fmri, kind="line", x="timepoint", y="signal", estimator=None
)
```

### Plotting subsets of data with semantic mappings

```python
sns.relplot(
    data=fmri, kind="line", x="timepoint", y="signal", hue="event"
)
```

```python
# Adding a style semantic to a line plot changes the pattern of dashes in the line by default:

sns.relplot(
    data=fmri, kind="line", x="timepoint", y="signal", hue="region", style="event"
)
```

```python
# we can also identify the subsets by the markers used at each observation:

sns.relplot(
    data=fmri, kind="line", x="timepoint", y="signal", hue="region", style="event", dashes=False, markers=True
)
```

```python
# The default colormap and handling of the legend in lineplot() also depends on whether the hue semantic is categorical 
# or numeric:

dots = sns.load_dataset("dots").query("align == 'dots'")
sns.relplot(
    data = dots, kind="line", x="time", y="firing_rate", hue="coherence", style="choice"
)
```

```python
# we can provide specific color values for each line by passing a list or dictionary:

palette = sns.cubehelix_palette(light=.8, n_colors=6)
sns.relplot(
    data=dots, kind="line", x="time", y="firing_rate", hue="coherence", style="choice", palette=palette,
)
```

```python
# we can also alter how the colormap is normalized:

from matplotlib.colors import LogNorm
palette = sns.cubehelix_palette(light=.7, n_colors=6)
sns.relplot(
    data=dots.query("coherence > 0"), kind="line", x="time", y="firing_rate", hue="coherence", style="choice",
    hue_norm=LogNorm(),
)
```

```python
# The third semantic, size, changes the width of the lines:

sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate",
    size="coherence", style="choice",
)
```

```python
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate",
    hue="coherence", size="choice", palette=palette,
)
```

### Controlling Sorting and orientation
    - The default behaviour of lineplot() is to sort the data by the x values before plotting, since it assumes you are most often trying to draw `y` as a function of `x`.

```python
healthexp = sns.load_dataset("healthexp").sort_values("Year")
sns.relplot(
    data=healthexp, kind="line",
    x="Spending_USD", y="Life_Expectancy", hue="Country",
    sort=False
)
```

```python
# Its also possible to sort or aggregate along the y axis:

sns.relplot(
    data=fmri, kind="line",
     x="signal", y="timepoint", hue="event",
    orient="y",
)
```

## Showing Multiple relationships with facets

```python
# The best approach for making multiple relationships is to make more than one plot:
sns.relplot(
    data = tips,
    x="total_bill", y="tip", hue="smoker", col="time"
)
```

```python
# we cn also show the influence of two variables by faceting on the columns and one by faceting on the rows:

sns.relplot(
    data=fmri, kind="line", x="timepoint", y="signal", hue="subject", col="region", row="event", height=3,
    estimator=None
)
```

```python
sns.relplot(
    data=fmri.query("region == 'frontal'"), kind="line",
    x="timepoint", y="signal", hue="event", style="event",
    col="subject", col_wrap=5,
    height=3, aspect=.75, linewidth=2.5,
)
```
