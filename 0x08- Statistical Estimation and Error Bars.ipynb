# Statistical Estimation and Error Bars

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_errorbars(arg, **kws):
    np.random.seed(sum(map(ord, "error_bars")))
    x = np.random.normal(0, 1, 100)
    f, axs = plt.subplots(2, figsize=(7, 2), sharex=True, layout="tight")
    sns.pointplot(x=x, errorbar=arg, **kws, capsize=.3, ax=axs[0])
    sns.stripplot(x=x, jitter=.3, ax=axs[1])
```

## Measures od Data Spread

### Standard  deviation error bars

```python
plot_errorbars("sd")
```

### Percentile interval error bars

```python
plot_errorbars(("pi", 50))
```

## Measures of Estimate Uncertainty

### Standasr error bars

```python
plot_errorbars("se")
```

### Confidence interval error bars
- The confidence interval is constructed by taking a percentile interval of the bootstrap distribution.
- By default `errorbars="ci"` draws a 95% confidence interval:

```python
plot_errorbars("ci")
```

```python
# we can scale the standard error, using the formula similar to the one mentioned above:

plot_errorbars(("se", 2))
```

```python
# The nonparametric bootstrap has advantages is similar to those of
# the percentile interval: it will naturally adapt to skewed and bounded data,
# in a way that a standard error interval cannot.
# It is also more general. While the standard error formula is specific to,
# to the mean, error bars can be computed using the bootstrap for any estimator.

plot_errorbars("ci", estimator="median")
```

```python
# Bootstraping involves randommness and the error bars will appear sliglty different each time you run the code:

plot_errorbars("ci", n_boot=5000, seed=10)
```

### Custom error bars
- It is possible t pass a generic function to the `errorbar` parameter. This function should take a vector and produce a apair of values representing the minimum and the maximum points of the interval:

```python
plot_errorbars(lambda x: (x.min(), x.max()))
```
